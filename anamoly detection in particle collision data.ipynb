```python
import pandas as pd

# URL for the dataset
url = "https://opendata.cern.ch/record/545/files/Dimuon_DoubleMu.csv"

# Load into DataFrame
df = pd.read_csv(url, low_memory=False)

# Show basic info
print("✅ Data loaded.")
print("Shape:", df.shape)
print("\nColumns:")
print(df.columns.tolist())

# Show first few rows
df.head()
```

```python
# Show summary statistics for numeric columns
print("\nSummary statistics:")
print(df.describe())

# Show count of missing values per column
print("\nMissing values per column:")
print(df.isna().sum().sort_values(ascending=False))
```

```python
# Example feature list — you may adjust based on actual column names
features = [
    'E1', 'pt1', 'eta1', 'phi1',
    'E2', 'pt2', 'eta2', 'phi2',
    'M'
]

# Filter only those features (drop rows with missing values in any of them)
df_phys = df[features].dropna().reset_index(drop=True)

print("✅ Physics-relevant subset shape:", df_phys.shape)
df_phys.head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X = scaler.fit_transform(df_phys)

print("✅ Features scaled. Sample of scaled data:")
print(X[:5])
```

```python
from sklearn.ensemble import IsolationForest
from sklearn.neighbors import LocalOutlierFactor

# Isolation Forest
iso = IsolationForest(n_estimators=200, contamination=0.02, random_state=42)
y_pred_iso = iso.fit_predict(X)
labels_iso = pd.Series(y_pred_iso).map({1: 'Normal', -1: 'Anomaly'})

# Local Outlier Factor
lof = LocalOutlierFactor(n_neighbors=20, contamination=0.02)
y_pred_lof = lof.fit_predict(X)
labels_lof = pd.Series(y_pred_lof).map({1: 'Normal', -1: 'Anomaly'})

# Append to dataframe
df_phys['anomaly_iso'] = labels_iso.values
df_phys['anomaly_lof'] = labels_lof.values

print("✅ Anomaly detection done. Counts (Isolation Forest):")
print(df_phys['anomaly_iso'].value_counts())
print("\nCounts (LOF):")
print(df_phys['anomaly_lof'].value_counts())
```

```python
import matplotlib.pyplot as plt

# Example: Plot pt1 vs pt2 coloured by Isolation Forest anomaly label
plt.figure(figsize=(8,6))
normal = df_phys[df_phys['anomaly_iso']=='Normal']
anomaly = df_phys[df_phys['anomaly_iso']=='Anomaly']

plt.scatter(normal['pt1'], normal['pt2'], c='blue', s=10, label='Normal')
plt.scatter(anomaly['pt1'], anomaly['pt2'], c='red', s=20, label='Anomaly')
plt.xlabel('pt1')
plt.ylabel('pt2')
plt.title('Anomaly detection (Isolation Forest) on Dimuon dataset')
plt.legend()
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(8,6))
sns.kdeplot(data=normal, x='M', label='Normal', color='blue')
sns.kdeplot(data=anomaly, x='M', label='Anomaly', color='red')
plt.xlabel('Invariant Mass (M)')
plt.ylabel('Density')
plt.title('Invariant Mass Distribution (KDE)')
plt.legend()
plt.show()
```
